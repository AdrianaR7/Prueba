# Introducción

# ¿Qué son las Bibliotecas en Python?
Las bibliotecas en Python son conjuntos de módulos, que son archivos que contienen funciones, clases y variables, diseñados para realizar tareas específicas.

Estas bibliotecas proporcionan funcionalidades predefinidas para resolver problemas comunes de manera eficiente, evitando que tengas que escribir código desde cero.

**Módulo**: Es un archivo con código en Python, que puede contener funciones, variables y clases.

**Biblioteca**: Una biblioteca es un conjunto de módulos que agrupan funcionalidades relacionadas. Por ejemplo, una biblioteca puede contener varios módulos para operaciones matemáticas, gráficos o manejo de fechas.


# Tipos de Bibliotecas en Python
**Bibliotecas** **Estándar**

Son las que vienen preinstaladas con Python.*Ejemplos* *Comunes*:

***math***: Funciones matemáticas básicas.

***datetime***: Funciones para manipular fechas y horas.

***os***: Funciones para interactuar con el sistema operativo.

**Bibliotecas** **de** **Terceros**

Son las que no forman parte de la instalación base de Python, pero que se instalan por separado. *Ejemplos*:

***numpy***: Para trabajar con arreglos y álgebra lineal.

***scipy***: Para operaciones científicas y numéricas avanzadas.

***pandas***: Para análisis y manipulación de datos tabulares.

# **math**

Se utiliza para realizar operaciones matemáticas básicas con números escalares.

```
print("EJEMPLO 1")
import math  # Se importa la librería 'math' para utilizar funciones matemáticas

x = 16  # Se asigna el valor 16 a la variable 'x'

resultado = math.sqrt(x)  # Se calcula la raíz cuadrada de 'x' usando 'math.sqrt' y se guarda en 'resultado'

print("La raíz cuadrada de 16 es:", resultado)  # Se imprime el resultado de la raíz cuadrada de 16
```

```
print("EJEMPLO 2")
import math  # Importamos la librería 'math'

base = 3  # Asignamos el valor 3 a la variable 'base'
exponente = 4  # Asignamos el valor 4 al exponente

resultado = math.pow(base, exponente)  # Calculamos 3 elevado a la potencia 4

print("3 elevado a la potencia de 4 es:", resultado)  # Imprimimos el resultado de la potencia
```

```
print("EJEMPLO 3")
import math  # Importamos la librería 'math'

numero = -15  # Asignamos el valor negativo -15 a la variable 'numero'

valor_absoluto = math.fabs(numero)  # Calculamos el valor absoluto de 'numero'

print("El valor absoluto de -15 es:",valor_absoluto)  # Imprimimos el valor absoluto
```

# **numpy**
Se utiliza para trabajar con arreglos y realizar operaciones matemáticas de manera eficiente.


```
import numpy as np  # Importamos la librería numpy y la renombramos como np

arr = np.array([1, 2, 3, 4])  # Creamos un arreglo de numpy con los elementos 1, 2, 3 y 4

half_values = arr / 2  # Calculamos la mitad de cada elemento del arreglo 'arr'

print("Las mitades de los elementos son:", half_values)  # Imprimimos el resultado
```

# **matplotlib**.**pyplot**

Se utiliza para crear gráficos y visualizar datos.


```
import matplotlib.pyplot as plt  # Importamos la librería matplotlib.pyplot para gráficos y la renombramos como plt

x = np.linspace(-10, 10, 100)
# Generamos 100 puntos desde -10 hasta 10 (espaciados uniformemente)

y = x**2
# Calculamos el valor de la función cuadrática f(x) = x^2 para cada valor de x

plt.plot(x, y)  # Graficamos los valores de x en el eje horizontal y los valores de y en el eje vertical

plt.title("Gráfico de f(x) = x^2")  # Añadimos un título al gráfico

plt.xlabel("x")  # Añadimos la etiqueta 'x' al eje horizontal

plt.ylabel("f(x)")  # Añadimos la etiqueta 'f(x)' al eje vertical

plt.grid(True)  # Activamos la cuadrícula en el gráfico

plt.show()  # Mostramos el gráfico en pantalla
```

# **scipy**
Permite realizar optimización y encontrar raíces de funciones.


```
import scipy.optimize as opt  # Importamos la librería 'scipy.optimize' con el alias 'opt', que contiene funciones para optimización

def f(x):  # Definimos una función f(x)
    return x**2 - 4  # Esta es la función f(x) = x^2 - 4, cuya raíz es 2

# Encontrar la raíz de la función usando el método de Newton
root = opt.newton(f, 1.8)
# Usamos 'opt.newton' para encontrar la raíz de f(x), con una estimación inicial de 1.8

print("La raíz encontrada es:", root)  # Imprimimos el resultado de la raíz encontrada
```

# **Tipos de Errores**


# **Error** **Absoluto**
Es la diferencia entre el valor exacto y el valor aproximado de una cantidad. Se mide en las mismas unidades que la cantidad que se está analizando.
Ejemplo: Si la solución exacta de un problema es 5.678 y la solución aproximada es 5.7, el error absoluto es ∣5.678-5.7∣=0.022.


```
valor_exacto = 5.678  # Asignamos el valor exacto (5.678) a la variable 'valor_exacto'

valor_aproximado = 5.7
# Asignamos el valor aproximado (5.7) a la variable 'valor_aproximado'

error_absoluto = abs(valor_exacto - valor_aproximado)
# Calculamos el error absoluto como la diferencia absoluta entre el valor exacto y el valor aproximado

print("Error absoluto:", error_absoluto)  # Imprimimos el error absoluto
```

# **Error Relativo**
Es el cociente del error absoluto entre el valor exacto. Mide la magnitud del error en relación con el valor exacto. Se expresa generalmente como un porcentaje.
Ejemplo: Usando los mismos valores de la sección anterior, el error relativo sería:
Error relativo=(∣5.678-5.7∣)/(5.678)=(0.022)/(5.678)≈0.0039 o 0.39%


```
# Definir el valor exacto y aproximado
valor_exacto = 5.678  # Asignamos el valor exacto (5.678) a la variable 'valor_exacto'
valor_aproximado = 5.7  # Asignamos el valor aproximado (5.7) a la variable 'valor_aproximado'

# Calcular el error absoluto
error_absoluto = abs(valor_exacto - valor_aproximado)
# Calculamos el error absoluto como la diferencia entre el valor exacto y el aproximado
print("Error absoluto:", error_absoluto)  # Imprimimos el error absoluto

# Calcular el error relativo (se debe hacer después de definir el error absoluto)
error_relativo = error_absoluto / valor_exacto
# Calculamos el error relativo, que es el error absoluto dividido entre el valor exacto
print("Error relativo:", error_relativo)  # Imprimimos el error relativo
```

# **Error** **de** **Redondeo**

Se produce debido a la representación finita de los números en las computadoras. Como las computadoras no pueden manejar decimales infinitos, redondean los números a la cantidad de dígitos que el sistema puede representar.
Ejemplo: Consideremos que intentamos almacenar el valor de π, que tiene una expansión decimal infinita. Sin embargo, se redondea a un número finito, como 3.14.


```
pi_exacto = math.pi
# Asignamos el valor exacto de pi (3.141592653589793) a la variable 'pi_exacto'

pi_redondeado = 3.14
# Asignamos un valor redondeado de pi (3.14) a la variable 'pi_redondeado'

error_redondeo = abs(pi_exacto - pi_redondeado)
# Calculamos el error de redondeo como la diferencia absoluta entre pi exacto y el valor redondeado

print("Error de redondeo:", error_redondeo)  # Imprimimos el error de redondeo
```


# **Error** **por** **Truncamiento**

Se produce cuando un proceso infinito o continuo es aproximado por un proceso finito. Esto es común cuando trabajamos con series infinitas o integrales, donde no podemos calcular todos los términos y debemos truncar el proceso en algún punto.
Ejemplo: Imagina que queremos calcular el valor de e (la base de los logaritmos naturales) de manera aproximada, usando la serie de Taylor:
e≈1+1/1!+1/2!+1/3!+⋯

Si truncamos esta serie después de unos pocos términos, estamos introduciendo un error de truncamiento.


```
import math  # Importamos la librería 'math' para utilizar funciones matemáticas

e_exacto = math.e
# Asignamos el valor exacto de la constante 'e' (Euler) a la variable 'e_exacto'

e_aproximado = 2.718
# Asignamos un valor truncado aproximado de 'e' a la variable 'e_aproximado'

error_truncamiento = abs(e_exacto - e_aproximado)
# Calculamos el error por truncamiento, que es la diferencia absoluta entre el valor exacto y el aproximado

print("Error por truncamiento:", error_truncamiento)  # Imprimimos el error por truncamiento
```
