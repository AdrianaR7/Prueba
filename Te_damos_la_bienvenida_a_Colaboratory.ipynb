# **scipy**
Permite realizar optimizacin y encontrar races de funciones.


```
import scipy.optimize as opt  # Importamos la librera 'scipy.optimize' con el alias 'opt', que contiene funciones para optimizacin

def f(x):  # Definimos una funcin f(x)
    return x**2 - 4  # Esta es la funcin f(x) = x^2 - 4, cuya raz es 2

# Encontrar la raz de la funcin usando el mtodo de Newton
root = opt.newton(f, 1.8)
# Usamos 'opt.newton' para encontrar la raz de f(x), con una estimacin inicial de 1.8

print("La raz encontrada es:", root)  # Imprimimos el resultado de la raz encontrada
```

# **Tipos de Errores**


# **Error** **Absoluto**
Es la diferencia entre el valor exacto y el valor aproximado de una cantidad. Se mide en las mismas unidades que la cantidad que se est analizando.
Ejemplo: Si la solucin exacta de un problema es 5.678 y la solucin aproximada es 5.7, el error absoluto es 5.678-5.7=0.022.


```
valor_exacto = 5.678  # Asignamos el valor exacto (5.678) a la variable 'valor_exacto'

valor_aproximado = 5.7
# Asignamos el valor aproximado (5.7) a la variable 'valor_aproximado'

error_absoluto = abs(valor_exacto - valor_aproximado)
# Calculamos el error absoluto como la diferencia absoluta entre el valor exacto y el valor aproximado

print("Error absoluto:", error_absoluto)  # Imprimimos el error absoluto
```

# **Error Relativo**
Es el cociente del error absoluto entre el valor exacto. Mide la magnitud del error en relacin con el valor exacto. Se expresa generalmente como un porcentaje.
Ejemplo: Usando los mismos valores de la seccin anterior, el error relativo sera:
Error relativo=(5.678-5.7)/(5.678)=(0.022)/(5.678)0.0039 o 0.39%


```
# Definir el valor exacto y aproximado
valor_exacto = 5.678  # Asignamos el valor exacto (5.678) a la variable 'valor_exacto'
valor_aproximado = 5.7  # Asignamos el valor aproximado (5.7) a la variable 'valor_aproximado'

# Calcular el error absoluto
error_absoluto = abs(valor_exacto - valor_aproximado)
# Calculamos el error absoluto como la diferencia entre el valor exacto y el aproximado
print("Error absoluto:", error_absoluto)  # Imprimimos el error absoluto

# Calcular el error relativo (se debe hacer despus de definir el error absoluto)
error_relativo = error_absoluto / valor_exacto
# Calculamos el error relativo, que es el error absoluto dividido entre el valor exacto
print("Error relativo:", error_relativo)  # Imprimimos el error relativo
```

# **Error** **de** **Redondeo**

Se produce debido a la representacin finita de los nmeros en las computadoras. Como las computadoras no pueden manejar decimales infinitos, redondean los nmeros a la cantidad de dgitos que el sistema puede representar.
Ejemplo: Consideremos que intentamos almacenar el valor de , que tiene una expansin decimal infinita. Sin embargo, se redondea a un nmero finito, como 3.14.


```
pi_exacto = math.pi
# Asignamos el valor exacto de pi (3.141592653589793) a la variable 'pi_exacto'

pi_redondeado = 3.14
# Asignamos un valor redondeado de pi (3.14) a la variable 'pi_redondeado'

error_redondeo = abs(pi_exacto - pi_redondeado)
# Calculamos el error de redondeo como la diferencia absoluta entre pi exacto y el valor redondeado

print("Error de redondeo:", error_redondeo)  # Imprimimos el error de redondeo
```


# **Error** **por** **Truncamiento**

Se produce cuando un proceso infinito o continuo es aproximado por un proceso finito. Esto es comn cuando trabajamos con series infinitas o integrales, donde no podemos calcular todos los trminos y debemos truncar el proceso en algn punto.
Ejemplo: Imagina que queremos calcular el valor de e (la base de los logaritmos naturales) de manera aproximada, usando la serie de Taylor:
e1+1/1!+1/2!+1/3!+

Si truncamos esta serie despus de unos pocos trminos, estamos introduciendo un error de truncamiento.


```
import math  # Importamos la librera 'math' para utilizar funciones matemticas

e_exacto = math.e
# Asignamos el valor exacto de la constante 'e' (Euler) a la variable 'e_exacto'

e_aproximado = 2.718
# Asignamos un valor truncado aproximado de 'e' a la variable 'e_aproximado'

error_truncamiento = abs(e_exacto - e_aproximado)
# Calculamos el error por truncamiento, que es la diferencia absoluta entre el valor exacto y el aproximado

print("Error por truncamiento:", error_truncamiento)  # Imprimimos el error por truncamiento
```
